```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "3"

import tensorflow as tf
import keras.backend.tensorflow_backend as KTF
 
config = tf.ConfigProto()  
config.gpu_options.allow_growth=True  
session = tf.Session(config=config)
KTF.set_session(session)
```

```python
from preprocessing import *
from models import *
from RobustFL import *
from sklearn.metrics import f1_score
```

```python
data_root_path =  '/data/qit16/MINIST/'
unlabeled_images, train_images, train_labels = load_data('train',data_root_path)
_, test_images, test_labels = load_data('test',data_root_path)
```

```python
train_users = local_data_partition(train_labels)
```

```python
trigger = TriggerGenerator(3,25,25,255,0,train_labels.shape[1])
train_images,train_labels,malicious_clients = generate_malicious_clients(train_users,train_images,train_labels,trigger)
poisoned_test_images,poisoned_test_labels = generate_poisoned_test_data(test_images,test_labels,trigger)
```

```python
lr = 0.05
user_num = 20
ckpt = 5

    
model = get_model(lr)
Result = []
DetectionScore = []

for i in range(500):
    p,r = RobustFL(i,user_num,model,unlabeled_images,train_images,train_labels,train_users,malicious_clients)
    if i%ckpt==0:
        pred = model.predict(test_images).argmax(axis=-1)
        lbs = test_labels.argmax(axis=-1)
        accuracy = f1_score(lbs,pred,average='micro')
        
        pred = model.predict(poisoned_test_images).argmax(axis=-1)
        lbs = poisoned_test_labels.argmax(axis=-1)
        DSR = 1-f1_score(lbs,pred,average='micro')

        Result.append([accuracy,DSR])

        print(i,accuracy,DSR)
```
